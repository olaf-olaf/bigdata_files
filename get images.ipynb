```python
def get_bing_json(search_term):
    subscription_key = "KEY"
    search_url = "https://api.cognitive.microsoft.com/bing/v7.0/images/search"
    headers = {"Ocp-Apim-Subscription-Key" : subscription_key , 'minWidth':"200", "minHeight":"200", "imageContent":"Portrait"}
    params  = {"q": search_term}
    response = requests.get(search_url, headers=headers, params=params)
    response.raise_for_status()
    search_results = response.json()
    return search_results

def get_image_from_json(json):
    images = []
    for item in json['value']:
#         print(item)
        try:
#             print(item['contentUrl'])
            response = requests.get(item['contentUrl'])
            print(response)
            image = Image.open(BytesIO(response.content))
            images.append(image)
        except: 
            print("ERROR")
            None
    return images
def write_images_to_directory(directory, images, query):
    counter = 1
    for image in images:
        if not image.mode == 'RGB':
            image = image.convert('RGB')
        image.save(directory+"/"+query+str(counter)+".jpeg")
        counter += 1;
```

```python
for search_term in ['cute dog', 'cute cat', 'cute duck', 'cute cow', 'cute rabbit', 'cute hamster', 'cute seal',
                 'cute elephant', 'cute squirrel', 'cute tiger']:
    print(search_term)
    json = get_bing_json(search_term)
#     print(json)
    images = get_image_from_json(json)
    write_images_to_directory('images/original', images, search_term)
print("DONE")
    
```
